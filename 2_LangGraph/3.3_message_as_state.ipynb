**Messages as State & Reducers**

- In a conversational agent, you dont want each node to completely override the message history. Instead append new messages. Hence LangGraph's Reducers come in handy

- Reducer functions allow us to specify how the state would be updated. If no reducer function is specified, it would update, override the key

- Define our MessageState with a special annotation.


```python
#!/usr/bin/python3
from typing import Annotated, TypedDict
from langchain_core.messages import AnyMessage
from langgraph.graph.message import add_messages

class MessageState(TypedDict):
    messages: Annotated[list[AnyMessage], add_messages]
# Tells langGraph to append messages as list!
```

The "add_messages" reducer ensures that when a node returns a messages key, its content is appended to the existing list, rather than replacing it.
